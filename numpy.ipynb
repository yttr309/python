# numpy를 이용한 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1,3,7,5]
# %matplotlib qt
plt.plot(y, color='red')
```

```python
y = [1,3,7,5]
x = [1,2,6,7]
# %matplotlib qt
plt.plot(x,y, color='red')
```

```python
x = [1,2,6,7]
y = [x*x for x in x]
# %matplotlib qt
plt.plot(x,y, 'b--', marker = 'o')
```

```python
x = [1,2,6,7]
y = [x*x for x in x]
# %matplotlib qt
plt.plot(x,y, 'b--', marker = 'o')
```

```python
a=list(range(1,10,5))
b=list(range(1,10,0.5))
```

```python
a=np.arange(-10,10,0.1)
a
```

```python
b=a*a
b
plt.plot(a,b, 'b--', marker = 'o', markersize=0)
```

```python
b2=10*a
plt.plot(a,b, 'r--', marker = 'o', markersize=0)
plt.plot(a,b2, 'b--', marker = 'o', markersize=0)
```

```python
b2=10*a
#%matplotlib qt
plt.figure(1)
plt.plot(a,b, 'r--', marker = 'o', markersize=0)
plt.figure(2)
plt.plot(a,b2, 'b--', marker = 'o', markersize=0)
```

```python
b2=10*a
plt.subplot(1,2,1)
plt.plot(a,b, 'r--', marker = 'o', markersize=0)
plt.subplot(1,2,2)
plt.plot(a,b2, 'b--', marker = 'o', markersize=0)
```

```python
t=np.arange(0,12,0.5)
y1=np.sin(t)
y2=np.cos(t)
plt.plot(t,y1, 'r--')
plt.plot(t,y2, 'g--', marker='*')
plt.xlabel('TIME')
plt.ylabel('AMP')
plt.legend(['LEG','END'])
plt.grid()
```

```python
%matplotlib qt
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1,2,1)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['aaa','bbb'])
plt.plot(x,y1,'r--',marker='*')
plt.plot(x,y2,'y--',marker='o')
plt.subplot(1,2,2)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y3,'o','y--')
plt.plot(x,y4,'-s','b--')
plt.legend(['aaa','bbb'])
```
